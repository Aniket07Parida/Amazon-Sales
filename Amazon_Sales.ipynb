```python
odisha_Data = df[df["State"] == "Odisha"]
odisha_Data
```

```python
plt.Figure(figsize=(20,12))
product_sales = df.groupby(df["Product_Category"])["Total_Sales_INR"].size()
plt.bar(product_sales.index,product_sales.values,color="Yellow")
plt.xlabel("Product_Category")
plt.ylabel("Sales_Number")
plt.title("Odisha Product Sales")
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
odisha_Data["Date"] = pd.to_datetime(odisha_Data["Date"])
filtered = odisha_Data[(odisha_Data["Date"] >= "2025-01-01") & (odisha_Data["Date"] <= "2025-05-31")]
filtered.set_index("Date",inplace=True)
month_of_sales = filtered["Total_Sales_INR"].resample("M").sum().sort_values(ascending=False)
plt.bar(month_of_sales.index.strftime("%b %Y"),
         month_of_sales.values)
plt.xlabel("Date")
plt.ylabel("Total_Sales")
plt.title("Sales of month")
plt.tight_layout()
plt.show()
```

```python
fig, axis = plt.subplots(2, 1, figsize=(12, 8))

# 1. Delivery Status (Pie)
total_delivery_status = df.groupby("Delivery_Status")["Total_Sales_INR"].sum()

axis[0].pie(
    total_delivery_status.values,
    labels=total_delivery_status.index,
    autopct='%1.1f%%',
    startangle=90,
    textprops={'fontsize': 10}
)
axis[0].set_title("Delivery Status Distribution", fontsize=14, fontweight='bold')

# Total sales per product category
total_product_sales = (
    df.groupby("Product_Category")["Total_Sales_INR"]
    .sum()
    .sort_values(ascending=False)  # optional: sort by sales
)

# Bar chart instead of histogram
axis[1].bar(
    total_product_sales.index,
    total_product_sales.values,
    edgecolor='black',
    color='green'
)

axis[1].set_title("Total Sales by Product Category", fontsize=14, fontweight='bold')
axis[1].set_xlabel("Product Category")
axis[1].set_ylabel("Total Sales (INR)")
axis[1].tick_params(axis='x', rotation=45)
axis[1].grid(axis='y', linestyle='--', alpha=0.5)


plt.show()
```

```python
fig, axis = plt.subplots(2, 1, figsize=(20, 15))

# Total Items Sales
total_item_sales = (
    df.groupby("Product_Name")["Total_Sales_INR"]
    .sum()
    .sort_values(ascending=True)
)

axis[0].barh(total_item_sales.index, total_item_sales.values,color="brown")
axis[0].set_title("Top 10 Products by Sales", fontsize=14, fontweight='bold')
axis[0].set_ylabel("Total Sales (INR)")
axis[0].tick_params(axis='x')
axis[0].grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

#Payment Method Salez
payment_status = (
    df.groupby("Payment_Method")["Total_Sales_INR"]
    .sum()
    .sort_values(ascending=True)
)

axis[1].plot(payment_status.index, payment_status.values, marker="o",color="red",linestyle="-.")
axis[1].set_title("Sales by Payment Method", fontsize=14, fontweight='bold')
axis[1].set_ylabel("Total Sales (INR)")
axis[1].tick_params(axis='x')
axis[1].grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split

# features and target
X = df[["Quantity"]]
y = df["Total_Sales_INR"]

# split (20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=48)

model = LinearRegression()
model.fit(X_train, y_train)

prediction = model.predict(X_test)

print("MAE:", mean_absolute_error(y_test, prediction))
print("R2 score:", r2_score(y_test, prediction))
```

```python
# Logistic Regression 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

X = df[["Quantity"]]
y = df["Total_Sales_INR"]


y_binary = (y > y.median()).astype(int) 
x_train, x_test, y_train, y_test = train_test_split(X, y_binary, train_size=0.2, random_state=48)

model = LogisticRegression()
model.fit(x_train, y_train)

prediction = model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, prediction))
print("Confusion Matrix:\n", confusion_matrix(y_test, prediction))
print("Classification Report:\n", classification_report(y_test, prediction))
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

X = df[["Product_Category"]]
y = df["Delivery_Status"]

# Encode categorical features to numerical
le_category = LabelEncoder()
X_encoded = X.copy()
X_encoded["Product_Category"] = le_category.fit_transform(X["Product_Category"])

le_status = LabelEncoder()
y_encoded = le_status.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, train_size=0.2, random_state=48)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```
