```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pyodbc 

conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=ANIKETPARIDA\\SQLEXPRESS03;"
    "DATABASE=Amazon_sales;"
    "Trusted_Connection=yes"
)
query = "SELECT * FROM amazon_sales"
df = pd.read_sql(query,conn)
df
```

```python
empty_row = df.isna().sum()
null_value = df.isnull().sum()
fill_empty = df.fillna(0)
fill_empty
```

```python
Total_Sales = df["Total_Sales_INR"].sum()
print(f"Total Sales of India {Total_Sales:.2f}")
```

```python
# Show percentages
category_percentage = df["Product_Category"].value_counts(normalize=True) * 100
print(category_percentage)
```

```python
odisha_Data = df[df["State"] == "Odisha"]
odisha_Data
```

```python
plt.Figure(figsize=(20,12))
product_sales = df.groupby(df["Product_Category"])["Total_Sales_INR"].size()
plt.bar(product_sales.index,product_sales.values,color="Yellow")
plt.xlabel("Product_Category")
plt.ylabel("Sales_Number")
plt.title("Odisha Product Sales")
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
product_items_sales = odisha_Data.groupby(odisha_Data["Product_Name"])["Total_Sales_INR"].sum()
product_items_sales
```

```python
odisha_Data["Date"] = pd.to_datetime(odisha_Data["Date"])
filtered = odisha_Data[(odisha_Data["Date"] >= "2025-01-01") & (odisha_Data["Date"] <= "2025-05-31")]
filtered.set_index("Date",inplace=True)
month_of_sales = filtered["Total_Sales_INR"].resample("M").sum().sort_values(ascending=False)
plt.bar(month_of_sales.index.strftime("%b %Y"),
         month_of_sales.values)
plt.xlabel("Date")
plt.ylabel("Total_Sales")
plt.title("Sales of month")
plt.tight_layout()
plt.show()
```

```python
# correct grouping
delivery_status = odisha_Data.groupby("Delivery_Status")["Total_Sales_INR"].sum()

# create figure and pie chart (sizes first, labels second)
plt.figure(figsize=(10,4))
plt.pie(delivery_status.values, labels=delivery_status.index, autopct="%1.1f%%")
plt.title("Sales by Delivery Status Odisha")
plt.tight_layout()
plt.show()
```

```python
sale_of_maharashra = df[df["State"] == "Maharashtra"]
sale_of_maharashra
```

```python
sale_of_maharashra["Date"] = pd.to_datetime(sale_of_maharashra["Date"])
jan_sales = sale_of_maharashra[(sale_of_maharashra["Date"] >= "2025-01-01") & (sale_of_maharashra["Date"] <= "2025-01-31")]
maha_jan = sale_of_maharashra.groupby("Date")["Total_Sales_INR"].sum()
maha_jan
```

```python
maximum_sales_product = sale_of_maharashra.groupby(sale_of_maharashra["Product_Category"])["Total_Sales_INR"].max()
maximum_sales_product
```

```python
average_product = sale_of_maharashra.groupby(sale_of_maharashra["Product_Name"])["Total_Sales_INR"].mean().sort_values(ascending=False)
plt.figure(figsize=(20, 9))
plt.bar(average_product.index, average_product.values, color="Orange")
plt.xlabel("Product Name")
plt.ylabel("Average Sales")
plt.title("Average Sales by Product")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
average_sales_state = df.groupby(df["State"])["Total_Sales_INR"].mean().sort_values(ascending=True)
plt.figure(figsize=(20,9))
plt.barh(average_sales_state.index,average_sales_state.values,color="green")
plt.xlabel("States")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45,ha="right")
plt.tight_layout()
plt.show()
```

```python
# Subplot of data 
fig, axis = plt.subplots(4, 1, figsize=(20, 8))

# Delivery Status
total_delivery_status = df.groupby(df["Delivery_Status"])["Total_Sales_INR"].sum()
axis[0].pie(total_delivery_status.values, labels=total_delivery_status.index, autopct='%1.1f%%', startangle=90)
axis[0].set_title("Delivery Status")

# Product Sales
total_product_sales = df.groupby(df["Product_Category"])["Total_Sales_INR"].sum().sort_values(ascending=False)
axis[1].bar(total_product_sales.index, total_product_sales.values, color="Green")
axis[1].set_xlabel("Product Category")
axis[1].set_ylabel("Total Sales")
axis[1].set_title("Total Product Sales")
axis[1].tick_params(axis='x', rotation=45)

# Total Item sales
total_item_sales = df.groupby(df["Product_Name"])["Total_Sales_INR"].sum().sort_values(ascending=False)
axis[2].bar(total_item_sales.index, total_item_sales.values, color="brown")
axis[2].set_xlabel("Product Name")
axis[2].set_ylabel("Total Sales")
axis[2].set_title("Product Sales")
axis[2].tick_params(axis='x', rotation=45)

# Total Payment method
payment_status = df.groupby(df["Payment_Method"])["Total_Sales_INR"].sum().sort_values(ascending=True)
axis[3].plot(payment_status.index, payment_status.values, marker="o", color="red")
axis[3].set_xlabel("Payment Method")
axis[3].set_ylabel("Total Payment")
axis[3].set_title("Total Payment Method")
axis[3].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
